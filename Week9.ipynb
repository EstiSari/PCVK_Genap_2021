```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
from skimage import io
from skimage import transform
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.morphology import (erosion, dilation, closing, opening,
                                area_closing, area_opening)
from skimage.color import rgb2gray
```

```python
img = cv.imread('/content/drive/MyDrive/pvck/Gambar/Images/j.png')

kernel = np.ones((5,5), np.uint8)
dilation = cv.dilate(img, kernel, iterations = 1)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.subplot(122),plt.imshow(dilation),plt.title('After Dilatation')
```

```python
img1= cv.imread('/content/drive/MyDrive/pvck/Gambar/Images/j.png', 0)

m,n= img1.shape 

plt.imshow(img1, cmap="gray")

k=5
SE= np.ones((k,k), dtype=np.uint8)
constant= (k-1)//2

imgErode= np.zeros((m,n), dtype=np.uint8)

for i in range(constant, m-constant):
  for j in range(constant,n-constant):
    temp= img1[i-constant:i+constant+1, j-constant:j+constant+1]
    product= temp*SE
    imgErode[i,j]= np.min(product)
plt.subplot(121), plt.imshow(img1, cmap ="gray"), plt.title('Original')
plt.subplot(122), plt.imshow(imgErode, cmap="gray"), plt.title('After Erotion')
```

```python
kernel = np.ones((5,5), np.uint8)
dilation = cv.morphologyEx(img, cv.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(dilation), plt.title('After Dilatation')
```

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv.morphologyEx(img, cv.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(erosion), plt.title('After Erotion')
```

```python
kernel = np.ones((7,7), np.uint8)
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(opening), plt.title('After Opening')
```

```python
kernel = np.ones((7,7), np.uint8)
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(closing), plt.title('After Closing')
```

```python
kernel = cv.getStructuringElement(cv.MORPH_CROSS, (3, 3))
cross3 = cv.morphologyEx(img, cv.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(cross3), plt.title('After Dilatation 3x3')
```

```python
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (3, 3))
cross3 = cv.morphologyEx(img, cv.MORPH_DILATE, kernel)
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(cross3), plt.title('After Circular 3x3')
```
